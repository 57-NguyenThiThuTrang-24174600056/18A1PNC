```python
class Stack:
    def __init__(self, max_size=10):
        self.stack = []
        self.max_size = max_size
        self.top = -1

    def __del__(self):
        print("Stack đã bị hủy.")

    def push(self, value):
        if self.isFull():
            print("Stack đầy! Không thể thêm phần tử.")
            return
        self.stack.append(float(value))
        self.top += 1
        print(f"Đã thêm {value} vào stack.")

    def pop(self):
        if self.isEmpty():
            print("Stack trống! Không thể lấy phần tử.")
            return None
        value = self.stack.pop()
        self.top -= 1
        print(f"Đã lấy {value} ra khỏi stack.")
        return value

    def isEmpty(self):
        return self.top == -1

    def isFull(self):
        return self.top == self.max_size - 1

    def count(self):
        return self.top + 1


# ===== Chương trình chính =====
if __name__ == "__main__":
    n = int(input("Nhập kích thước tối đa của stack: "))
    s = Stack(n)

    while True:
        print("\n1. Thêm phần tử (push)")
        print("2. Lấy phần tử (pop)")
        print("3. Kiểm tra rỗng")
        print("4. Kiểm tra đầy")
        print("5. Đếm số phần tử")
        print("6. Thoát")

        chon = input("Chọn thao tác: ")

        if chon == "1":
            value = input("Nhập giá trị cần thêm: ")
            s.push(value)
        elif chon == "2":
            s.pop()
        elif chon == "3":
            print("Stack rỗng." if s.isEmpty() else "Stack không rỗng.")
        elif chon == "4":
            print("Stack đầy." if s.isFull() else "Stack chưa đầy.")
        elif chon == "5":
            print("Số phần tử trong stack:", s.count())
        elif chon == "6":
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ.")
```
