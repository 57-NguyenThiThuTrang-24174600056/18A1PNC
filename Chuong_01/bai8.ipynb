```python
from typing import List

class Date:
    def __init__(self, day=1, month=1, year=1900):
        self.day = day
        self.month = month
        self.year = year

    def display(self):
        print(f"{self.day:02d}/{self.month:02d}/{self.year}", end="")

    def next(self):
        days_in_month = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
        if self.month == 2 and self.year % 4 == 0 and (self.year % 100 != 0 or self.year % 400 == 0):
            days_in_month[2] = 29
        if self.day < days_in_month[self.month]:
            self.day += 1
        else:
            self.day = 1
            if self.month < 12:
                self.month += 1
            else:
                self.month = 1
                self.year += 1

class Employee:
    def __init__(self, name: str, birth_date: Date, join_date: Date):
        self.name = name
        self.birth_date = birth_date
        self.join_date = join_date

    def display(self):
        print(f"Tên: {self.name}")
        print("Ngày sinh: ", end="")
        self.birth_date.display()
        print()
        print("Ngày vào công ty: ", end="")
        self.join_date.display()
        print()

def main():
    employees: List[Employee] = []
    while True:
        print("\n--- Quản lý nhân viên ---")
        print("1. Thêm nhân viên")
        print("2. Hiển thị tất cả")
        print("3. Xóa nhân viên (theo tên)")
        print("4. Thoát")
        choice = input("Chọn: ")
        if choice == "1":
            name = input("Tên: ")
            bd_day = int(input("Ngày sinh (ngày): "))
            bd_month = int(input("Tháng: "))
            bd_year = int(input("Năm: "))
            jd_day = int(input("Ngày vào công ty (ngày): "))
            jd_month = int(input("Tháng: "))
            jd_year = int(input("Năm: "))
            birth = Date(bd_day, bd_month, bd_year)
            join = Date(jd_day, jd_month, jd_year)
            emp = Employee(name, birth, join)
            employees.append(emp)
            print("Đã thêm!")
        elif choice == "2":
            if not employees:
                print("Không có nhân viên.")
            else:
                for emp in employees:
                    emp.display()
        elif choice == "3":
            name = input("Tên cần xóa: ")
            employees = [emp for emp in employees if emp.name != name]
            print("Đã xóa (nếu tồn tại).")
        elif choice == "4":
            break

if __name__ == "__main__":
    main()
```
