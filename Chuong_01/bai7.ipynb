```python
class Date:
    def __init__(self, day=1, month=1, year=2000):
        self.day = day
        self.month = month
        self.year = year
        self._validate_date()
    
    def _is_leap_year(self):
        """Ki·ªÉm tra nƒÉm nhu·∫≠n"""
        return (self.year % 4 == 0 and self.year % 100 != 0) or (self.year % 400 == 0)
    
    def _days_in_month(self):
        """Tr·∫£ v·ªÅ s·ªë ng√†y trong th√°ng"""
        days_in_month = [31, 29 if self._is_leap_year() else 28, 31, 30, 31, 30,
                         31, 31, 30, 31, 30, 31]
        return days_in_month[self.month - 1]
    
    def _validate_date(self):
        """Ki·ªÉm tra v√† ƒëi·ªÅu ch·ªânh ng√†y/th√°ng h·ª£p l·ªá"""
        if self.month < 1 or self.month > 12:
            self.month = 1
        max_days = self._days_in_month()
        if self.day < 1 or self.day > max_days:
            self.day = 1
    
    def display(self):
        """In ra ng√†y/th√°ng/nƒÉm"""
        print(f"{self.day:02d}/{self.month:02d}/{self.year}")
    
    def next(self):
        """T√≠nh ng√†y k·∫ø ti·∫øp"""
        max_days = self._days_in_month()
        
        if self.day < max_days:
            self.day += 1
        else:
            self.day = 1
            if self.month < 12:
                self.month += 1
            else:
                self.month = 1
                self.year += 1
        return Date(self.day, self.month, self.year)


# ===== PH·∫¶N CH·∫†Y TH·ª¨ =====
print("\n=== TEST DATE ===")
# Nh·∫≠p theo ƒë·ªãnh d·∫°ng: dd mm yyyy
day, month, year = map(int, input("Nh·∫≠p ng√†y th√°ng nƒÉm (c√°ch nhau b·∫±ng kho·∫£ng tr·∫Øng): ").split())

# T·∫°o ƒë·ªëi t∆∞·ª£ng Date
date1 = Date(day, month, year)

print("üìÖ Ng√†y hi·ªán t·∫°i:")
date1.display()

print("üìÜ Ng√†y ti·∫øp theo:")
date1.next().display()
```
