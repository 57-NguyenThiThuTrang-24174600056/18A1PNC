```python
class Stack:
    def __init__(self, max_size=10):
        self.stack = []
        self.max_size = max_size

    def __del__(self):
        print("Stack đã bị hủy.")

    def push(self, value):
        if self.isFull():
            print("Stack đầy! Không thể thêm phần tử.")
            return
        try:
            self.stack.append(float(value))
            print(f"Đã thêm {value} vào stack.")
        except ValueError:
            print("Giá trị không hợp lệ! Vui lòng nhập số.")

    def pop(self):
        if self.isEmpty():
            print("Stack trống! Không thể lấy phần tử.")
            return None
        value = self.stack.pop()
        print(f"Đã lấy {value} ra khỏi stack.")
        return value

    def isEmpty(self):
        return len(self.stack) == 0

    def isFull(self):
        return len(self.stack) >= self.max_size

# ===== Chương trình chính =====
if __name__ == "__main__":
    n = int(input("Nhập kích thước tối đa của stack: "))
    s = Stack(n)

    while True:
        print("\n1. Push (thêm phần tử)")
        print("2. Pop (lấy phần tử)")
        print("3. Kiểm tra rỗng")
        print("4. Kiểm tra đầy")
        print("5. Thoát")

        chon = input("Chọn thao tác: ")

        if chon == "1":
            value = input("Nhập giá trị cần thêm: ")
            s.push(value)
        elif chon == "2":
            s.pop()
        elif chon == "3":
            if s.isEmpty():
                print("Stack đang rỗng.")
            else:
                print("Stack không rỗng.")
        elif chon == "4":
            if s.isFull():
                print("Stack đã đầy.")
            else:
                print("Stack chưa đầy.")
        elif chon == "5":
            print("Kết thúc chương trình.")
            break
        else:
            print("Lựa chọn không hợp lệ.")



```
