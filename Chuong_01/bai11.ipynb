```python
import math

# =================== LỚP TAM GIÁC ===================
class TamGiac:
    def __init__(self, a=0, b=0, c=0):
        self.a = a
        self.b = b
        self.c = c

    def nhap(self):
        print("\n=== Nhập thông tin tam giác thường ===")
        self.a = float(input("Nhập cạnh a: "))
        self.b = float(input("Nhập cạnh b: "))
        self.c = float(input("Nhập cạnh c: "))

    def chu_vi(self):
        return self.a + self.b + self.c

    def dien_tich(self):
        s = self.chu_vi() / 2
        return math.sqrt(s * (s - self.a) * (s - self.b) * (s - self.c))

    def hien_thi(self):
        print(f"Tam giác có 3 cạnh ({self.a}, {self.b}, {self.c})")


# =================== TAM GIÁC VUÔNG ===================
class TamGiacVuong(TamGiac):
    def __init__(self, a=0, b=0):
        c = math.sqrt(a**2 + b**2)
        super().__init__(a, b, c)

    def nhap(self):
        print("\n=== Nhập thông tin tam giác vuông ===")
        a = float(input("Nhập cạnh góc vuông thứ nhất (a): "))
        b = float(input("Nhập cạnh góc vuông thứ hai (b): "))
        self.__init__(a, b)  # Cập nhật lại c

    def hien_thi(self):
        print(f"Tam giác vuông có 2 cạnh góc vuông ({self.a}, {self.b}), cạnh huyền = {self.c:.2f}")


# =================== TAM GIÁC CÂN ===================
class TamGiacCan(TamGiac):
    def __init__(self, a=0, day=0):  # Hai cạnh bằng nhau
        super().__init__(a, a, day)

    def nhap(self):
        print("\n=== Nhập thông tin tam giác cân ===")
        a = float(input("Nhập cạnh bên (a): "))
        day = float(input("Nhập cạnh đáy: "))
        self.__init__(a, day)

    def hien_thi(self):
        print(f"Tam giác cân có 2 cạnh bên = {self.a}, cạnh đáy = {self.c}")


# =================== TAM GIÁC ĐỀU ===================
class TamGiacDeu(TamGiacCan):
    def __init__(self, a=0):
        super().__init__(a, a)

    def nhap(self):
        print("\n=== Nhập thông tin tam giác đều ===")
        a = float(input("Nhập độ dài cạnh: "))
        self.__init__(a)

    def dien_tich(self):
        return (math.sqrt(3) / 4) * self.a ** 2

    def hien_thi(self):
        print(f"Tam giác đều có cạnh = {self.a}")


# =================== CHƯƠNG TRÌNH CHÍNH ===================
def main():
    print("=== CHƯƠNG TRÌNH TÍNH CHU VI & DIỆN TÍCH CÁC LOẠI TAM GIÁC ===")
    print("1. Tam giác thường")
    print("2. Tam giác vuông")
    print("3. Tam giác cân")
    print("4. Tam giác đều")

    chon = input("Nhập lựa chọn (1-4): ")

    if chon == "1":
        tg = TamGiac()
        tg.nhap()
    elif chon == "2":
        tg = TamGiacVuong()
        tg.nhap()
    elif chon == "3":
        tg = TamGiacCan()
        tg.nhap()
    elif chon == "4":
        tg = TamGiacDeu()
        tg.nhap()
    else:
        print("Lựa chọn không hợp lệ!")
        return

    print("\n--- KẾT QUẢ ---")
    tg.hien_thi()
    print(f"Chu vi = {tg.chu_vi():.2f}")
    print(f"Diện tích = {tg.dien_tich():.2f}")


# =================== CHẠY CHƯƠNG TRÌNH ===================
if __name__ == "__main__":
    main()
```
