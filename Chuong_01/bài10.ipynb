```python
import math

# =================== LỚP CƠ SỞ: ĐIỂM ===================
class Diem:
    def __init__(self, x=0, y=0):
        self.x = x
        self.y = y
    
    def nhap_diem(self):
        """Nhập tọa độ điểm"""
        self.x = float(input("Nhập tọa độ x: "))
        self.y = float(input("Nhập tọa độ y: "))
    
    def hien_thi(self):
        """Hiển thị tọa độ điểm"""
        print(f"Tọa độ tâm: ({self.x}, {self.y})")


# =================== LỚP ELIP ===================
class Elip(Diem):
    def __init__(self, x=0, y=0, truc_lon=0, truc_nho=0):
        super().__init__(x, y)
        self.truc_lon = truc_lon
        self.truc_nho = truc_nho
    
    def nhap_elip(self):
        print("\n=== Nhập thông tin Elip ===")
        self.nhap_diem()
        self.truc_lon = float(input("Nhập độ dài trục lớn (a): "))
        self.truc_nho = float(input("Nhập độ dài trục nhỏ (b): "))
    
    def tinh_dien_tich(self):
        """Diện tích elip = π * a * b"""
        return math.pi * self.truc_lon * self.truc_nho
    
    def hien_thi(self):
        print(f"Elip có tâm tại ({self.x}, {self.y}), trục lớn = {self.truc_lon}, trục nhỏ = {self.truc_nho}")


# =================== LỚP ĐƯỜNG TRÒN ===================
class DuongTron(Elip):
    def __init__(self, x=0, y=0, ban_kinh=0):
        super().__init__(x, y)
        self.ban_kinh = ban_kinh
        self.truc_lon = self.truc_nho = ban_kinh * 2  # Đường tròn là elip đặc biệt
    
    def nhap_duong_tron(self):
        print("\n=== Nhập thông tin Đường tròn ===")
        self.nhap_diem()
        self.ban_kinh = float(input("Nhập bán kính: "))
        self.truc_lon = self.truc_nho = self.ban_kinh * 2
    
    def tinh_dien_tich(self):
        """Diện tích đường tròn = π * r^2"""
        return math.pi * self.ban_kinh ** 2
    
    def hien_thi(self):
        print(f"Đường tròn có tâm tại ({self.x}, {self.y}), bán kính = {self.ban_kinh}")


# =================== CHƯƠNG TRÌNH CHÍNH ===================
def main():
    print("=== TÍNH DIỆN TÍCH ELIP VÀ ĐƯỜNG TRÒN ===")

    # Nhập Elip
    elip = Elip()
    elip.nhap_elip()
    elip.hien_thi()
    print(f"➡ Diện tích elip: {elip.tinh_dien_tich():.2f}")

    # Nhập Đường tròn
    tron = DuongTron()
    tron.nhap_duong_tron()
    tron.hien_thi()
    print(f"➡ Diện tích đường tròn: {tron.tinh_dien_tich():.2f}")


# =================== CHẠY CHƯƠNG TRÌNH ===================
if __name__ == "__main__":
    main()
```
