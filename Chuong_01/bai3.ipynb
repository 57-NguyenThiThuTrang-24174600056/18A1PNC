```python
class PhanSo:
    def __init__(self, tu=0, mau=1): 
        if self.kiem_tra_hop_le(tu, mau):
            self.tu = tu
            self.mau = mau
        else:
            self.tu = 0
            self.mau = 1
            print("Phân số không hợp lệ, dùng mặc định 0/1")
    
    def kiem_tra_hop_le(self, tu, mau):  # Check mau != 0
        return mau != 0
    
    def nhap_phan_so(self):  # Nhập từ bàn phím
        self.tu = int(input("Nhập tử số: "))
        self.mau = int(input("Nhập mẫu số: "))
        if not self.kiem_tra_hop_le(self.tu, self.mau):
            print("Mẫu số không hợp lệ! Dùng 0/1")
            self.tu, self.mau = 0, 1
    
    def in_phan_so(self):  # In dạng tu/mau
        print(f"{self.tu}/{self.mau}")
ps = PhanSo()
ps.nhap_phan_so()
ps.in_phan_so()
```
