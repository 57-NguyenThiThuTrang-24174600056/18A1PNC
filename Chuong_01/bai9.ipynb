```python
import math

# =================== L·ªöP C∆† S·ªû: ƒêA GI√ÅC ===================
class DaGiac:
    def __init__(self, so_canh: int, do_dai_canh: float):
        self.so_canh = so_canh
        self.do_dai_canh = do_dai_canh

    def tinh_chu_vi(self):
        return self.so_canh * self.do_dai_canh

    def tinh_dien_tich(self):
        """T√≠nh di·ªán t√≠ch ƒëa gi√°c ƒë·ªÅu"""
        if self.so_canh < 3:
            return 0
        return (self.so_canh * self.do_dai_canh**2) / (4 * math.tan(math.pi / self.so_canh))

    def __str__(self):
        return f"ƒêa gi√°c {self.so_canh} c·∫°nh: Chu vi = {self.tinh_chu_vi():.2f}, Di·ªán t√≠ch = {self.tinh_dien_tich():.2f}"


# =================== L·ªöP H√åNH B√åNH H√ÄNH ===================
class HinhBinhHanh:
    def __init__(self, canh_day: float, canh_ben: float, goc: float):  # goc t√≠nh theo ƒë·ªô
        self.canh_day = canh_day
        self.canh_ben = canh_ben
        self.goc = math.radians(goc)  # chuy·ªÉn sang radian ƒë·ªÉ t√≠nh to√°n

    def tinh_chu_vi(self):
        return 2 * (self.canh_day + self.canh_ben)

    def tinh_dien_tich(self):
        return self.canh_day * self.canh_ben * math.sin(self.goc)

    def __str__(self):
        return f"H√¨nh b√¨nh h√†nh: Chu vi = {self.tinh_chu_vi():.2f}, Di·ªán t√≠ch = {self.tinh_dien_tich():.2f}"


# =================== L·ªöP H√åNH CH·ªÆ NH·∫¨T ===================
class HinhChuNhat(HinhBinhHanh):
    def __init__(self, chieu_dai: float, chieu_rong: float):
        super().__init__(chieu_dai, chieu_rong, 90)  # g√≥c vu√¥ng m·∫∑c ƒë·ªãnh

    def tinh_dien_tich(self):
        return self.canh_day * self.canh_ben

    def __str__(self):
        return f"H√¨nh ch·ªØ nh·∫≠t: Chu vi = {self.tinh_chu_vi():.2f}, Di·ªán t√≠ch = {self.tinh_dien_tich():.2f}"


# =================== L·ªöP H√åNH VU√îNG ===================
class HinhVuong(HinhChuNhat):
    def __init__(self, canh: float):
        super().__init__(canh, canh)

    def tinh_dien_tich(self):
        return self.canh_day ** 2

    def __str__(self):
        return f"H√¨nh vu√¥ng: Chu vi = {self.tinh_chu_vi():.2f}, Di·ªán t√≠ch = {self.tinh_dien_tich():.2f}"


# =================== CH∆Ø∆†NG TR√åNH CH√çNH ===================
def main():
    while True:
        print("\n========== MENU H√åNH H·ªåC ==========")
        print("1. ƒêa gi√°c ƒë·ªÅu")
        print("2. H√¨nh b√¨nh h√†nh")
        print("3. H√¨nh ch·ªØ nh·∫≠t")
        print("4. H√¨nh vu√¥ng")
        print("0. Tho√°t ch∆∞∆°ng tr√¨nh")
        lua_chon = input("üëâ M·ªùi b·∫°n ch·ªçn h√¨nh c·∫ßn t√≠nh: ")

        if lua_chon == "1":
            n = int(input("Nh·∫≠p s·ªë c·∫°nh c·ªßa ƒëa gi√°c: "))
            a = float(input("Nh·∫≠p ƒë·ªô d√†i m·ªói c·∫°nh: "))
            hinh = DaGiac(n, a)
        elif lua_chon == "2":
            a = float(input("Nh·∫≠p ƒë·ªô d√†i c·∫°nh ƒë√°y: "))
            b = float(input("Nh·∫≠p ƒë·ªô d√†i c·∫°nh b√™n: "))
            goc = float(input("Nh·∫≠p g√≥c gi·ªØa hai c·∫°nh (ƒë·ªô): "))
            hinh = HinhBinhHanh(a, b, goc)
        elif lua_chon == "3":
            dai = float(input("Nh·∫≠p chi·ªÅu d√†i: "))
            rong = float(input("Nh·∫≠p chi·ªÅu r·ªông: "))
            hinh = HinhChuNhat(dai, rong)
        elif lua_chon == "4":
            canh = float(input("Nh·∫≠p ƒë·ªô d√†i c·∫°nh: "))
            hinh = HinhVuong(canh)
        elif lua_chon == "0":
            print("üëã C·∫£m ∆°n b·∫°n ƒë√£ s·ª≠ d·ª•ng ch∆∞∆°ng tr√¨nh! T·∫°m bi·ªát!")
            break
        else:
            print("‚ö†Ô∏è L·ª±a ch·ªçn kh√¥ng h·ª£p l·ªá, vui l√≤ng nh·∫≠p l·∫°i.")
            continue

        print("‚û° K·∫øt qu·∫£:", hinh)


# =================== CH·∫†Y CH∆Ø∆†NG TR√åNH ===================
if __name__ == "__main__":
    main()

```
