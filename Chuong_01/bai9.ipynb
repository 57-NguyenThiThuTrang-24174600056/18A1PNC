```python
import math

# =================== Lá»šP CÆ  Sá»: ÄA GIÃC ===================
class DaGiac:
    def __init__(self, so_canh: int, do_dai_canh: float):
        self.so_canh = so_canh
        self.do_dai_canh = do_dai_canh

    def tinh_chu_vi(self):
        return self.so_canh * self.do_dai_canh

    def tinh_dien_tich(self):
        """TÃ­nh diá»‡n tÃ­ch Ä‘a giÃ¡c Ä‘á»u"""
        if self.so_canh < 3:
            return 0
        return (self.so_canh * self.do_dai_canh**2) / (4 * math.tan(math.pi / self.so_canh))

    def __str__(self):
        return f"Äa giÃ¡c {self.so_canh} cáº¡nh: Chu vi = {self.tinh_chu_vi():.2f}, Diá»‡n tÃ­ch = {self.tinh_dien_tich():.2f}"


# =================== Lá»šP HÃŒNH BÃŒNH HÃ€NH ===================
class HinhBinhHanh:
    def __init__(self, canh_day: float, canh_ben: float, goc: float):  # goc tÃ­nh theo Ä‘á»™
        self.canh_day = canh_day
        self.canh_ben = canh_ben
        self.goc = math.radians(goc)  # chuyá»ƒn sang radian Ä‘á»ƒ tÃ­nh toÃ¡n

    def tinh_chu_vi(self):
        return 2 * (self.canh_day + self.canh_ben)

    def tinh_dien_tich(self):
        return self.canh_day * self.canh_ben * math.sin(self.goc)

    def __str__(self):
        return f"HÃ¬nh bÃ¬nh hÃ nh: Chu vi = {self.tinh_chu_vi():.2f}, Diá»‡n tÃ­ch = {self.tinh_dien_tich():.2f}"


# =================== Lá»šP HÃŒNH CHá»® NHáº¬T ===================
class HinhChuNhat(HinhBinhHanh):
    def __init__(self, chieu_dai: float, chieu_rong: float):
        super().__init__(chieu_dai, chieu_rong, 90)  # gÃ³c vuÃ´ng máº·c Ä‘á»‹nh

    def tinh_dien_tich(self):
        return self.canh_day * self.canh_ben

    def __str__(self):
        return f"HÃ¬nh chá»¯ nháº­t: Chu vi = {self.tinh_chu_vi():.2f}, Diá»‡n tÃ­ch = {self.tinh_dien_tich():.2f}"


# =================== Lá»šP HÃŒNH VUÃ”NG ===================
class HinhVuong(HinhChuNhat):
    def __init__(self, canh: float):
        super().__init__(canh, canh)

    def tinh_dien_tich(self):
        return self.canh_day ** 2

    def __str__(self):
        return f"HÃ¬nh vuÃ´ng: Chu vi = {self.tinh_chu_vi():.2f}, Diá»‡n tÃ­ch = {self.tinh_dien_tich():.2f}"


# =================== CHÆ¯Æ NG TRÃŒNH CHÃNH ===================
def main():
    while True:
        print("\n========== MENU HÃŒNH Há»ŒC ==========")
        print("1. Äa giÃ¡c Ä‘á»u")
        print("2. HÃ¬nh bÃ¬nh hÃ nh")
        print("3. HÃ¬nh chá»¯ nháº­t")
        print("4. HÃ¬nh vuÃ´ng")
        print("0. ThoÃ¡t chÆ°Æ¡ng trÃ¬nh")
        lua_chon = input("ğŸ‘‰ Má»i báº¡n chá»n hÃ¬nh cáº§n tÃ­nh: ")

        if lua_chon == "1":
            n = int(input("Nháº­p sá»‘ cáº¡nh cá»§a Ä‘a giÃ¡c: "))
            a = float(input("Nháº­p Ä‘á»™ dÃ i má»—i cáº¡nh: "))
            hinh = DaGiac(n, a)
        elif lua_chon == "2":
            a = float(input("Nháº­p Ä‘á»™ dÃ i cáº¡nh Ä‘Ã¡y: "))
            b = float(input("Nháº­p Ä‘á»™ dÃ i cáº¡nh bÃªn: "))
            goc = float(input("Nháº­p gÃ³c giá»¯a hai cáº¡nh (Ä‘á»™): "))
            hinh = HinhBinhHanh(a, b, goc)
        elif lua_chon == "3":
            dai = float(input("Nháº­p chiá»u dÃ i: "))
            rong = float(input("Nháº­p chiá»u rá»™ng: "))
            hinh = HinhChuNhat(dai, rong)
        elif lua_chon == "4":
            canh = float(input("Nháº­p Ä‘á»™ dÃ i cáº¡nh: "))
            hinh = HinhVuong(canh)
        elif lua_chon == "0":
            print("ğŸ‘‹ Cáº£m Æ¡n báº¡n Ä‘Ã£ sá»­ dá»¥ng chÆ°Æ¡ng trÃ¬nh! Táº¡m biá»‡t!")
            break
        else:
            print("âš ï¸ Lá»±a chá»n khÃ´ng há»£p lá»‡, vui lÃ²ng nháº­p láº¡i.")
            continue

        print("â¡ Káº¿t quáº£:", hinh)


# =================== CHáº Y CHÆ¯Æ NG TRÃŒNH ===================
if __name__ == "__main__":
    main()

```
