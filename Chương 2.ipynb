Bài 2.2:

```python
import xml.etree.ElementTree as ET
from xml.dom import minidom

# --- Tạo root ---
students = ET.Element("Students")

# Danh sách sinh viên
student_list = [
    {"id": "SV001", "FullName": "Nguyễn Văn A", "YearOfBirth": "2006", "Class": "DHKL18A1HN", "Gender": "Nam"},
    {"id": "SV002", "FullName": "Trần Thị B", "YearOfBirth": "2006", "Class": "DHKL18A2HN", "Gender": "Nữ"},
    {"id": "SV003", "FullName": "Lê Văn C", "YearOfBirth": "2006", "Class": "DHKL18A3HN", "Gender": "Nam"}
]

# --- Sinh node XML ---
for s in student_list:
    student = ET.SubElement(students, "Student", id=s["id"])
    ET.SubElement(student, "FullName").text = s["FullName"]
    ET.SubElement(student, "YearOfBirth").text = s["YearOfBirth"]
    ET.SubElement(student, "Class").text = s["Class"]
    ET.SubElement(student, "Gender").text = s["Gender"]

# --- Ghi file XML ---
tree = ET.ElementTree(students)
tree.write("students.xml", encoding="utf-8", xml_declaration=True)

# --- Hiển thị XML đẹp trong notebook ---
xml_str = ET.tostring(students, encoding='utf-8')
pretty_xml = minidom.parseString(xml_str).toprettyxml(indent="  ")
print(pretty_xml)
```

```python
import requests
import xml.etree.ElementTree as ET
import csv

# URL RSS feed
url = "http://www.hindustantimes.com/rss/topnews/rssfeed.xml"

# Tải và lưu file RSS
response = requests.get(url)
with open("rssfeed.xml", "wb") as f:
    f.write(response.content)

# Phân tích XML
tree = ET.parse("rssfeed.xml")
root = tree.getroot()

# Namespace thường có trong RSS, cần xử lý
ns = {'ns': 'http://purl.org/rss/1.0/'}

# Trích xuất tin tức
items = root.findall(".//item")
news_list = []

for item in items:
    title = item.findtext("title")
    link = item.findtext("link")
    pubDate = item.findtext("pubDate")
    news_list.append({"title": title, "link": link, "pubDate": pubDate})

# Ghi ra file CSV
with open("news.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["title", "link", "pubDate"])
    writer.writeheader()
    writer.writerows(news_list)

print("Đã lưu tin tức vào file news.csv")
```

Bài 2.10:

```python
import json

# ----- Bước 1: Tạo dữ liệu JSON mô phỏng -----
company_data = {
    "company_name": "Công ty ABC",
    "address": "Hà Nội",
    "departments": {
        "Đơn vị A1": [
            {"name": "A"},
            {"name": "B"},
            {"name": "C"}
        ],
        "Đơn vị A2": [
            {"name": "D"},
            {"name": "E"}
        ],
        "Đơn vị A3": [
            {"name": "F"},
            {"name": "G"},
            {"name": "H"},
            {"name": "I"}
        ],
        "Đơn vị A4": [
            {"name": "K"}
        ],
        "Đơn vị A5": [
            {"name": "L"},
            {"name": "M"}
        ]
    }
}

# ----- Bước 2: Lưu dữ liệu vào file JSON -----
with open("company.json", "w", encoding="utf-8") as f:
    json.dump(company_data, f, ensure_ascii=False, indent=4)

# ----- Bước 3: Đọc dữ liệu từ file JSON -----
with open("company.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# ----- Bước 4: Thống kê nhân viên -----
company_name = data["company_name"]
address = data["address"]
departments = data["departments"]

# Tổng số nhân viên
total_employees = sum(len(staff) for staff in departments.values())

# ----- Bước 5: In kết quả -----
print("Kết quả thực hiện chương trình:")
print(f"Tên công ty: {company_name}")
print(f"Địa chỉ: {address}")
print("----- Thống kê nhân viên theo đơn vị -----")

for i, (dept_name, staff_list) in enumerate(departments.items(), start=1):
    count = len(staff_list)
    percent = (count / total_employees) * 100
    print(f"{i}. Tên đơn vị: {dept_name}")
    print(f"   - Số nhân viên: {count}")
    print(f"   - Tỷ lệ: {percent:.2f}%")
```

Bài 2.11:

```python
import json
from datetime import datetime

transactions = []

while True:
    print("\n--- Nhập thông tin giao dịch ---")
    loai = input("Loại giao dịch (Tiền/Vàng): ")
    so_luong = input("Số lượng: ")
    gia = input("Đơn giá: ")

    transactions.append({
        "loai": loai,
        "so_luong": so_luong,
        "gia": gia
    })

    tiep = input("Tiếp tục nhập? (y/n): ").lower()
    if tiep != 'y':
        break

# Khi người dùng không muốn nhập nữa
chon = input("\nBạn có muốn ghi vào tệp không? (1: Có / 0: Không): ")

if chon == "1":
    now = datetime.now().strftime("%Y-%m-%d-%H-%M-%S")
    filename = f"{now}.json"

    with open(filename, "w", encoding="utf-8") as f:
        json.dump(transactions, f, ensure_ascii=False, indent=4)

    print(f"Dữ liệu đã được ghi vào tệp: {filename}")
else:
    print("Không ghi dữ liệu vào tệp.")
```

```python
import requests

url = "https://jsonplaceholder.typicode.com/posts"

# Gửi yêu cầu GET đến API
response = requests.get(url)
posts = response.json()

# Tổng số bài post
print("Tổng số bài post:", len(posts))
print("\n--- Danh sách các bài post ---")

for post in posts[:5]:  # chỉ in 5 bài đầu cho gọn
    print(f"userId: {post['userId']}, id: {post['id']}")
    print(f"title: {post['title']}")
    print(f"body: {post['body']}\n")
```
